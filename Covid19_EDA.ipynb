### Import important libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

### Read the dataset

```python
df = pd.read_csv("Covid19.csv")
```

### Check basic information of the dataset

```python
df.head()
```

```python
df.info()
```

### **Basic Questions**
### Which Country has maximum total cases?

```python
df_confirmed=df.sort_values(by=['Confirmed'],ascending=False)
df_confirmed.head()
```

```python
fig=px.bar(df_confirmed.iloc[0:20],color='Country/Region',y='Confirmed',x='Country/Region',template="plotly_dark",title="<b>Top 20 countries with the most confirmed cases</b>")
fig.show()
```

### Which Country has maximum total deaths?

```python
df_death=df.sort_values(by=['Deaths'],ascending=False)
df_death.head()
```

```python
fig=px.bar(df_death.iloc[0:20],color='Country/Region',y='Deaths',x='Country/Region',template="plotly_dark",title="<b>Top 20 countries with the most deaths</b>")
fig.show()
```

### Which Country has most active cases?

```python
df_active=df.sort_values(by=['Active'],ascending=False)
df_active.head()
```

```python
fig=px.bar(df_active.iloc[0:20],color='Country/Region',y='Active',x='Country/Region',template="plotly_dark",title="<b>Top 20 countries with the most active cases</b>")
fig.show()
```

### Which Country has highest death rate?

```python
df_death_connfirmed=df.sort_values(by=['Deaths / 100 Cases'],ascending=False)
df_death_connfirmed.head()
```

```python
fig=px.bar(df_death_connfirmed.iloc[0:20],color='Country/Region',y='Deaths / 100 Cases',x='Country/Region',template="plotly_dark",title="<b>Top 20 countries with the highest death rate</b>")
fig.show()
```

### Pie Chart Representation of stats of countries of confirmed cases

```python
labels=df_confirmed[0:20]['Country/Region'].values
fig=px.pie(df_confirmed[0:20],values='Confirmed',names=labels,template="plotly_dark",title="Confirmed cases distribution of countries".format('Country/Region'))
fig.show()
```

### Pie Chart Representation of stats of countries of deaths

```python
labels=df_death[0:20]['Country/Region'].values
fig=px.pie(df_death[0:20],values='Deaths',names=labels,template="plotly_dark",title="Death distribution of countries".format('Country/Region'))
fig.show()
```
